```python
from openai import OpenAI
from google.colab import userdata

client = OpenAI(api_key=userdata.get('OpenAI_API_key'))
```

1.Generate Image
-Dalle2,3


```python
response = client.images.generate(
    model="dall-e-2",
    prompt="a cat wearing sunglass",
    size="256x256",
    n=1
)

# print(response.data[0].url)
```

```python
from IPython.display import Image

display(Image(url=response.data[0].url)) #print the image from url to this page
```

2.model= galle-3


```python
response = client.images.generate(
    model="dall-e-3",
    prompt="a cat wearing sunglasses",
    size="512x512",
    quality="standard",
    n=1
)
```

```python
display(Image(url=response.data[0].url))
```

Variation (Dalle2 only)

```python
response = client.images.create_variation(
  image=open("/content/cat.png", "rb"),
  n=2, #will generate different amount based of number of n
  size="256x256" #have to match image size
)

display(Image(url=response.data[0].url))
```

Edit (Dalle2 only)

```python
response = client.images.edit(
  image=open("/content/cat.png", "rb"),
  mask=open("/content/mask.png", "rb"),
  prompt="a cat wearing sunglasses at moon background", #by adding prompt to edit the mask photo, the Ai will compare the mask and ori image in order to edit the mask photo based on prompt command
  n=1,
  size = "256x256"
  )

display(Image(url=response.data[0].url))
```
