```python
from openai import OpenAI
from google.colab import userdata
from IPython.display import Image

client = OpenAI(api_key=userdata.get('OpenAI_API_key'))
```

#Storybook

Text model:
-Story=> Characters, plot, themes, setting, etc.

Image model:
-Cover


```python
#Story
def story_gen(prompt):
  system_prompt = """
  You are a world renownwd author for young adults fiction short stories.
  Given a concept, generate a short story to the themes of the concept with a twist ending.
  """

  response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
      {"role": "system", "content": system_prompt},
      {"role":"user","content":prompt}
  ],
    temperature = 1,
    max_tokens=2000
  )

  return response.choices[0].message.content
```

```python
prompt = "an old men with the sea"
print(story_gen(prompt))
```

```python
#cover art
from IPython.display import Image


def art_gen(prompt):
  response = client.images.generate(
      model = 'dall-e-2',
      prompt = prompt,
      size = '256x256',
      n=1
  )

  return response.data[0].url
```

```python
# cover prompt design
def design_gen(prompt):
  system_prompt = """
  You will be given a short story. Generate a prompt for a cover art that is suitable for the story. The prompt is for dallas-e-2.
  """
  response = client.chat.completions.create(
      model = 'gpt-4o-mini',
      messages = [
          {'role':'system', 'content':system_prompt},
          {'role':'user', 'content':prompt}
      ],
      temperature = 1.3,
      max_tokens = 1000
  )

  return response.choices[0].message.content
```

```python
prompt = "an old men with the sea"
story = story_gen(prompt)
design = design_gen(story)
art = art_gen(design)
display(Image(url = art))
print(design)
print(story)
```
